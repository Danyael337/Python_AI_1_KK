### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
1. The Object-Oriented Approach (Recommended)
    This method involves creating figure and axes objects explicitly and then calling methods on these objects.
    It provides more flexibility and is the preferred approach for complex plots.

2. The Pyplot (State-Based) Approach
    Uses global functions from matplotlib.pyplot that work like MATLAB-style plotting.
    Suitable for quick, simple plots, but can get messy in complex visualizations.

The Object-Oriented Approach is the preferred method, especially for advanced or multi-plot figures. 
The Pyplot Approach is fine for quick plots but becomes harder to manage in complex cases.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure - The entire canvas where plots are drawn.
         Can contain multiple subplots (axes).
         Example: fig = plt.figure()

Axes - Represents a single plot inside the figure.
       Contains x-axis, y-axis, labels, ticks, and lines.
       Example: ax = fig.add_subplot(1,1,1)

Axis - One-dimensional part of a plot (either x-axis or y-axis).
       Example: ax.set_xlim(0, 10)

Artist - Any element drawn on a figure (text, lines, markers, etc.).
         Example: ax.plot(x, y) adds an artist.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
In Matplotlib, the expected input data type for plotting functions like plt.plot() is typically array-like, 
which includes:
-NumPy arrays (np.array([1, 2, 3]))
-Python lists ([1, 2, 3])
-Pandas Series (df["column_name"])
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# OBJECT ORIENTED approach. 

import numpy as np
import matplotlib.pyplot as plt

# Generate x values using np.linspace
x = np.linspace(-4, 4, 100)
y = x**2  # Compute y = x^2

# Create figure and axis objects
fig, ax = plt.subplots()

# Plot the function
ax.plot(x, y, label="$y = x^2$", color="blue")

# Set title and labels
ax.set_title("Quadratic Function ($y = x^2$)")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

# Add grid and legend
ax.grid(True)
ax.legend()

# Show the plot
plt.show()
```

```python
# Pyplot approach. 

import numpy as np
import matplotlib.pyplot as plt

# Generate x and y values
x = np.linspace(-4, 4, 100)
y = x**2

# Plot using Pyplot
plt.plot(x, y, label="$y = x^2$", color="red")

# Add title and axis labels
plt.title("Quadratic Function ($y = x^2$)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
# Generate data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Generate data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot in the first subplot
axes[0].scatter(random_data_x, random_data_y, alpha=0.5, color="blue")
axes[0].plot(x, y, color="red", label="$y = x^2$")
axes[0].set_title("Scatter Plot with Quadratic Function")
axes[0].set_xlabel("Random X Data")
axes[0].set_ylabel("Random Y Data")
axes[0].grid(True)
axes[0].legend()

# Bar plot in the second subplot
axes[1].bar(names, values, color=["purple", "red", "orange", "yellow", "green"])
axes[1].set_title("Fruit Quantity")
axes[1].set_xlabel("Fruit Type")
axes[1].set_ylabel("Quantity")
axes[1].grid(axis="y")

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
